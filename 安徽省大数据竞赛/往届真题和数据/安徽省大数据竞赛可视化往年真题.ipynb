```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn
import numpy
import pandas as pd
```

```python
df = pd.read_csv("./Titanic.csv")
```

```python
Survied = df.Pclass[df.Survived==1].value_counts()
Died = df.Pclass[df.Survived==0].value_counts()
DF=pd.DataFrame({'huojiu':Survied,'yunan':Died})
DF.plot(kind='bar',stacked= True)
plt.title("huojiu_count")
plt.xlabel("level")
plt.ylabel("number")
plt.show()
```

```python
males = (df.Sex =='male').sum()
females = (df.Sex =='female').sum()
plt.pie([males,females],
       autopct='%.1f%%',
       labels=['males','females'])
plt.axis('equal')
plt.title('sex proportion')
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig = plt.figure(figsize=(15,8))
fig.set(alpha=0.2)

plt.subplot2grid((2,3),(0,0))
df.Survived.value_counts().plot(kind='bar')
plt.title('huojiu yunan')
plt.ylabel('number')

plt.subplot2grid((2,3),(0,1))
df.Pclass.value_counts().plot(kind='bar')
plt.title('level')
plt.ylabel('count')

plt.subplot2grid((2,3),(0,2))
plt.scatter(df.Survived,df.Age)
plt.ylabel('age')
plt.grid(b=True,which='major',axis='y')
plt.title('Age (1:huojiu)')

plt.subplot2grid((2,3),(1,0),colspan=2)
Pclass1 = df.Age[df.Pclass==1].value_counts()
Pclass2 = df.Age[df.Pclass==2].value_counts()
Pclass3 = df.Age[df.Pclass==3].value_counts()
DF1 = pd.DataFrame({'1':Pclass1,'2':Pclass2,'3':Pclass3})
plt.plot(DF1)
plt.xlabel('age')
plt.ylabel('midu')
plt.title('level')
plt.legend(('1','2','3'),loc='best')

plt.subplot2grid((2,3),(1,2))
df.Embarked.value_counts().plot(kind='bar')
plt.title('load_count')
plt.ylabel('count')
plt.show()
```

```python
students = pd.read_csv("./students.csv")
```

```python
students.columns=['course','name','score']
```

```python
math = students.score[students.course =='math'].mean()
computer = students.score[students.course =='computer'].mean()
english = students.score[students.course =='english'].mean()
X = ['computer','math','english']
y = [computer,math,english]
plt.bar(X,y)
plt.title('pingjunchengji')
plt.xlabel('course')
plt.ylabel('score')
plt.show()
```

```python
math = students[students.course =='math'][['name','score']]
english = students[students.course =='english'][['name','score']]
computer = students[students.course =='computer'][['name','score']]
algorithm = students[students.course =='algorithm'][['name','score']]
```

```python
fig = plt.figure(figsize=(15,8))
plt.plot(math['name'],math['score'])
plt.plot(computer['name'],computer['score'])
plt.plot(english['name'],english['score'])
plt.plot(algorithm['name'],algorithm['score'])
plt.legend(('math','computer','english','algorithm'),loc='best')
plt.xlabel('course')
plt.ylabel('score')
plt.xticks(rotation=90)
plt.title('name-course-score')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
math = students[students.course=='math'][['name','score']].sort_values(by='name',ascending=True)
english = students[students.course=='english'][['name','score']].sort_values(by='name',ascending=True)
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.bar(english['name'],english['score'])
ax2 = ax1.twinx()
ax2.plot(math['name'],math['score'],c='r')
ax2.legend(('english'),loc='best')
ax1.legend(('math'),loc='best')
plt.xticks(rotation=90)
plt.show()
```

```python
pm = pd.read_csv('./airpm25.txt',sep=',')
pm.columns=['date','num','city','area']
pm.head()
```

```python

hefei = pm['num'].loc[pm.city=='合肥'].mean()
huangshan = pm['num'].loc[pm.city=='黄山'].mean()
wuhu = pm['num'].loc[pm.city=='芜湖'].mean()
```

```python
X = ['合肥','黄山','芜湖']
y = [hefei,huangshan,wuhu]
plt.bar(X,y)
plt.title('三市平均PM')
plt.xlabel('城市')
plt.ylabel('数值')
plt.savefig("1.png")
plt.show()
```

```python
hefei1 = pm.loc[pm.city=='合肥'].groupby('date').mean()
huangshan1 = pm.loc[pm.city=='黄山'].groupby('date').mean()
wuhu1 = pm.loc[pm.city=='芜湖'].groupby('date').mean()
```

```python
fig = plt.figure(figsize=(15,8))
x=['2018-12-01','2018-12-02','2018-12-03','2018-12-04','2018-12-05','2018-12-06','2018-12-07','2018-12-08','2018-12-09','2018-12-10']
plt.plot(x,hefei1,marker='o',c='r')
plt.plot(x,huangshan1,marker='o',c='g')
plt.plot(x,wuhu1,marker='o',c='b')
plt.legend(['合肥','黄山','芜湖'],loc='best')
plt.title('三市PM随日期变化')
plt.xlabel('日期')
plt.ylabel('数值')
plt.savefig("2.png")
plt.show()
```

```python
gaoxin=pm['num'].loc[(pm.city=='合肥')&(pm.area=='高新区')]
luyang=pm['num'].loc[(pm.city=='合肥')&(pm.area=='庐阳区')]
list_num= ['2018-12-01','2018-12-02','2018-12-03','2018-12-04','2018-12-05','2018-12-06','2018-12-07','2018-12-08','2018-12-09','2018-12-10']

fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(111)
ax1.bar(list_num,gaoxin,tick_label=list_num,label='高新区')
ax1.set_ylabel('高新区PM2.5')

ax2= ax1.twinx()
ax2.plot(list_num,luyang,c='r',marker='o')
ax2.set_ylabel('庐阳区PM2.5')
plt.title('高新区和庐阳区PM2.5随日期的变化图')
plt.savefig("3.png")
plt.show()
```
