```python
import matplotlib 
import matplotlib.pyplot as plt
import pandas as pd

order = pd.read_csv("./Order.csv")
order.columns=['order_ID','date','area','product_ID','kind','sell','num','discount','profit']

sell_2015 = order.loc[order['date'].str.contains('2015')]
sell_2015_all = sell_2015['sell'].sum()

sell_2016 = order.loc[order['date'].str.contains('2016')]
sell_2016_all = sell_2016['sell'].sum()

sell_2017 = order.loc[order['date'].str.contains('2017')]
sell_2017_all = sell_2017['sell'].sum()

sell_2018 = order.loc[order['date'].str.contains('2018')]
sell_2018_all = sell_2018['sell'].sum()


X=['2015','2016','2017','2018']
y=[sell_2015_all,sell_2016_all,sell_2017_all,sell_2018_all]
plt.rcParams['font.sans-serif']=['SimHei']
fig=plt.figure(figsize=(15,8))
plt.bar(X,y)
plt.title('年销售额变化')
plt.xlabel('年份')
plt.ylabel('数量')
plt.show()
```

```python
list_sell=[sell_2015_all,sell_2016_all,sell_2017_all,sell_2018_all]
list_profit=[profit_2015_all,profit_2016_all,profit_2017_all,profit_2018_all]

plt.rcParams['font.sans-serif']=['SimHei']
fig=plt.figure(figsize=(15,8))
plt.title('销售额与利润关系')
plt.plot(list_sell,list_profit,marker='o',c='r')
plt.xlabel('年销量(单位：百万)')
plt.ylabel('年利润')
plt.show
```

```python
bangong = order.loc[order['kind'].str.contains('办公用品')][['area','sell']]
area_bangong_all=bangong.sell.groupby(bangong['area']).sum()
jiaju = order.loc[order['kind'].str.contains('家具')][['area','sell']]
area_jiaju_all=jiaju.sell.groupby(jiaju['area']).sum()
jishu = order.loc[order['kind'].str.contains('技术')][['area','sell']]
area_jishu_all=jishu.sell.groupby(jishu['area']).sum()
area_jiaju_all.sort_values()
```
